```python
def get_residual_sum_of_sqaures(model,data,outcome):
    prediction=model.predict(data)
    residuals=outcome-prediction
    
    RSS=(residuals*residuals).sum()
    return RSS
```

```python
rss11=get_residual_sum_of_sqaures(model1,train_data[model_1_features],train_data['price'])
rss22=get_residual_sum_of_sqaures(model2,train_data[model_2_features],train_data['price'])
rss33=get_residual_sum_of_sqaures(model3,train_data[model_3_features],train_data['price'])


print(rss1)
print(rss2)
print(rss3)
```
